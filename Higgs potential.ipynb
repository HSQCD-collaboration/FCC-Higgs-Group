## Higgs Potential
$$ V(\phi) \propto - \phi^2 + \frac{1}{2} \phi^4 $$

```python
import numpy as np
import matplotlib.pyplot as plt

phi = np.linspace(-2, 2, 100)
vPhi = -(phi**2) + (1/2)*(phi**4)  # Corrected the expression for vPhi

plt.plot(phi, vPhi)  # Corrected variable name from vphi to vPhi
plt.xlabel('phi')
plt.ylabel('vPhi')
plt.title('Plot of vPhi vs phi')
plt.grid(True)
plt.show()
```
